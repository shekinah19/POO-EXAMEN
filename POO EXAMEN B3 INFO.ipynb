```python
plt.figure()
plt.hist(df['quality'], bins=10)
plt.xlabel("Qualité")
plt.ylabel("Fréquence")
plt.title("Distribution de la qualité du vin")
plt.show()
```

```python
plt.figure()
plt.imshow(df.corr())
plt.colorbar()
plt.title("Matrice de corrélation")
plt.show()
```

```python
plt.figure()
plt.boxplot([df[df['quality']==q]['alcohol'] for q in sorted(df['quality'].unique())])
plt.xlabel("Qualité")
plt.ylabel("Alcohol")
plt.title("Distribution de l'alcool selon la qualité")
plt.show()
```

```python
plt.figure()
plt.scatter(df['alcohol'],df['quality']) 
plt.xlabel("Alcohol")
plt.ylabel("Qualité")
plt.title("Relation entre alcool et qualité")
plt.show()
```

```python
quality_counts = df['quality'].value_counts().sort_index()

plt.figure()
plt.bar(quality_counts.index,quality_counts.values)
plt.xlabel("Qualité")
plt.ylabel("Nombre de vins")
plt.title("Nombre de vins par qualité")
plt.show()
```
